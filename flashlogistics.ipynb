# 🚚 FlashLogistics — Calculadora de Envíos

Sistema para calcular el coste final de un envío según la distancia en km.

---
### 🐛 Bugs corregidos:
| Función | Bug original | Corrección |
|---|---|---|
| `calcular_coste_distancia` | `return` vacío (devolvía `None`) | `return precio` |
| `aplicar_descuento` | Devolvía el descuento, no el total restante | `return total - descuento` |

```python
# ── Funciones ──────────────────────────────────────────────────────────────

def calcular_coste_distancia(km):
    """Calcula el coste base del envío: 0.50€ por km."""
    precio = km * 0.5
    return precio  # ✅ Bug 1 corregido: antes era 'return' vacío


def aplicar_descuento(total):
    """Aplica un descuento del 20% si el total supera los 100€."""
    if total > 100:
        descuento = total * 0.20
        return total - descuento  # ✅ Bug 2 corregido: antes devolvía solo 'descuento'
    return total


def mostrar_ticket(km, coste_base, final):
    """Imprime el resumen del envío."""
    print("="*40)
    print("       🚚 TICKET DE ENVÍO - FlashLogistics")
    print("="*40)
    print(f"  Distancia:       {km} km")
    print(f"  Coste base:      {coste_base:.2f}€")
    if final < coste_base:
        print(f"  Descuento (20%): -{coste_base - final:.2f}€")
    print("-"*40)
    print(f"  TOTAL:           {final:.2f}€")
    print("="*40)
```

```python
# ── Programa Principal con input ────────────────────────────────────────────

try:
    distancia = float(input("Introduce la distancia del envío en km: "))

    if distancia <= 0:
        print("❌ La distancia debe ser un número positivo.")
    else:
        coste_base = calcular_coste_distancia(distancia)
        total_final = aplicar_descuento(coste_base)
        mostrar_ticket(distancia, coste_base, total_final)

except ValueError:
    print("❌ Error: introduce un número válido (ej: 150 o 75.5)")
```
