## ✈️ Ejemplo 1 — Control de Altitud

### ¿Qué hace este código?

Simula un sistema de monitoreo de altitud para un dron o aeronave. Se define un **límite de seguridad** de 500 metros y se comprueban altitudes desde 0 hasta 1000 m en escalones de 100 m.

### Conceptos clave

| Elemento | Descripción |
|---|---|
| `range(0, 1001, 100)` | Genera los valores: 0, 100, 200, ..., 1000 |
| `if metros > altitud_maxima` | Compara cada valor contra el límite |
| f-string `f"..{metros}m"` | Inserta variables dentro de un texto |

### Flujo del programa

```
metros = 0, 100, 200, 300, 400, 500  →  Altitud controlada ✅
metros = 600, 700, 800, 900, 1000    →  ¡Alerta! ⚠️
```

````python
# ─────────────────────────────────────────────
# EJEMPLO 1: Control de Altitud
# ─────────────────────────────────────────────

altitud_maxima = 500   # Límite de seguridad (metros)
incremento = 100       # Paso entre cada lectura

for metros in range(0, 1001, incremento):
    if metros > altitud_maxima:
        print(f"Alerta: {metros}m excede el limite de seguridad.")
    else:
        print(f"Altitud controlada: {metros}m.")
````

- ¿Qué pasa si cambias el paso a `incremento = 50`? ¿Cuántas iteraciones ejecuta el bucle?
- Prueba a cambiar `>` por `>=` y observa cómo afecta al valor exacto de 500 m.

## 📦 Ejemplo 2 — Inventario de Suministros

### ¿Qué hace este código?

Simula el consumo de suministros de una misión. Parte de **10 unidades** y descuenta 3 en cada turno. Cuando las unidades bajan de 4, activa **una única alerta** de nivel crítico.

### Conceptos clave

| Elemento | Descripción |
|---|---|
| `while unidades > 0` | Repite el bloque mientras haya stock |
| `unidades -= 3` | Equivale a `unidades = unidades - 3` |
| `critico = True` | Bandera booleana para disparar la alerta solo una vez |
| `not critico` | Solo entra al `if` si la alerta aún no fue activada |

### Traza de ejecución

```
Inicio: unidades = 10
  → imprime 10, resta 3 → 7
  → imprime 7,  resta 3 → 4
  → imprime 4,  resta 3 → 1  ← aquí unidades <= 4, activa alerta
  → imprime 1,  resta 3 → -2 ← while se detiene (no > 0)
Fin: imprime mensaje de inventario agotado
```

````python
# ─────────────────────────────────────────────
# EJEMPLO 2: Inventario de Suministros
# ─────────────────────────────────────────────

unidades = 10
critico = False   # Bandera: ¿ya se emitió la alerta crítica?

while unidades > 0:
    print(f"Suministros restantes: {unidades}")
    unidades -= 3
    if not critico and unidades <= 4:
        print("¡Aviso! Nivel crítico detectado.")
        critico = True   # Evita repetir el aviso

print("Inventario agotado o insuficiente para la próxima misión.")
````

### 💡 Reflexión

- ¿Qué ocurre si eliminas la bandera `critico` y la condición `not critico`? ¿Cuántas veces aparece el aviso?
- ¿Qué pasa si cambias el descuento de 3 a 1? ¿Cuántas iteraciones más ejecuta el bucle?
- El programa termina cuando `unidades` es **negativa** (−2). ¿Cómo lo mejorarías para que nunca baje de 0?

## 🔬 Ejemplo 3 — Análisis de Muestras con `break`

### ¿Qué hace este código?

Procesa una lista de lecturas de sensores. Si encuentra un **valor negativo** (dato inválido), interrumpe el análisis inmediatamente y reporta cuántas muestras se procesaron correctamente antes del error.

### Conceptos clave

| Elemento | Descripción |
|---|---|
| `for valor in muestras` | Recorre cada elemento de la lista |
| `break` | Interrumpe el bucle al instante, sin procesar el resto |
| `procesadas += 1` | Contador que solo avanza si el valor es válido |

### Visualización de la lista

```
Índice:   0     1     2      3      4     5
Valor:  [1.2,  0.8,  3.5,  -0.1,  2.2,  1.5]
          ✅    ✅    ✅     ❌ STOP
                            ↑
                     break aquí → procesadas = 3
```

````python
# ─────────────────────────────────────────────
# EJEMPLO 3: Análisis de Muestras con break
# ─────────────────────────────────────────────

muestras = [1.2, 0.8, 3.5, -0.1, 2.2, 1.5]
procesadas = 0

for valor in muestras:
    if valor < 0:
        print(f"Error: Muestra negativa detectada ({valor}). Abortando escaneo.")
        break          # Sale del bucle inmediatamente
    procesadas += 1    # Solo se ejecuta si valor >= 0

print(f"Se han analizado {procesadas} muestras correctamente.")
````

### 💡 Reflexión

- ¿Qué pasaría si mueves el `-0.1` al primer lugar de la lista? ¿Cuántas muestras se procesarían?
- ¿Qué pasa si **no hay ningún valor negativo** en la lista? El bucle termina normalmente sin ejecutar el `break`.
- ¿Cómo cambiarías el código para que, en vez de abortar, simplemente **salte** la muestra negativa y continúe? (Pista: usa `continue` en lugar de `break`.)

## 📡 Ejemplo 4 — Panel de Sensores (Diccionario)

### ¿Qué hace este código?

Recorre un **diccionario** de lecturas de sensores y compara cada valor contra un límite de alerta. Si alguna lectura lo supera, imprime un aviso de peligro; si no, confirma que el sensor está en rango.

### Conceptos clave

| Elemento | Descripción |
|---|---|
| `{"clave": valor}` | Diccionario: pares nombre → dato |
| `for nombre in sensores` | Itera sobre las **claves** del diccionario |
| `sensores[nombre]` | Accede al **valor** asociado a esa clave |

### Estado de los sensores

```
Sensor       Lectura   Límite   Estado
──────────────────────────────────────
Oxígeno        98       100      ✅ OK
Presión       105       100      ❌ PELIGRO
Radiación      12       100      ✅ OK
Humedad        40       100      ✅ OK
```

````python
# ─────────────────────────────────────────────
# EJEMPLO 4: Panel de Sensores
# ─────────────────────────────────────────────

sensores = {
    "Oxígeno":   98,
    "Presión":  105,
    "Radiación": 12,
    "Humedad":   40
}

limite_alerta = 100

for nombre in sensores:
    lectura = sensores[nombre]
    if lectura > limite_alerta:
        print(f"PELIGRO: {nombre} fuera de rango ({lectura}).")
    else:
        print(f"Sensor {nombre}: OK.")
````
