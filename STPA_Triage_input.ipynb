# 🏥 Sistema de Triaje y Protocolo de Aislamiento (STPA)

Software para personal de enfermería que clasifica pacientes según su gravedad y determina si deben ser aislados por riesgo biológico.

---

```python
# ============================================================
#   INGRESO DE DATOS DEL PACIENTE
# ============================================================

print("=" * 45)
print("   🏥  INGRESO DE DATOS DEL PACIENTE")
print("=" * 45)

# Datos numéricos
edad                = int(input("Edad (años): "))
frecuencia_cardiaca = int(input("Frecuencia cardíaca (bpm): "))
temperatura         = float(input("Temperatura (°C): "))

# Datos booleanos — el usuario escribe 's' o 'n'
tiene_sintomas_infecciosos = input("¿Tiene síntomas infecciosos? (s/n): ").strip().lower() == "s"
esta_vacunado              = input("¿Está vacunado? (s/n): ").strip().lower() == "s"
es_emergencia_critica      = input("¿Es emergencia crítica? (s/n): ").strip().lower() == "s"

print("\n✅ Datos registrados.")
```

```python
# ============================================================
#   CLASIFICACIÓN DE PRIORIDAD (if / elif / else)
# ============================================================

# 1. Prioridad ROJA: emergencia crítica O temperatura > 41 °C
if es_emergencia_critica or temperatura > 41:
    prioridad, nivel, emoji_prioridad = "ROJA", "Inmediata", "🔴"

# 2. Prioridad NARANJA: fiebre >= 39 °C Y taquicardia > 100 bpm
elif temperatura >= 39 and frecuencia_cardiaca > 100:
    prioridad, nivel, emoji_prioridad = "NARANJA", "Urgente", "🟠"

# 3. Prioridad AMARILLA: menor de 12 O mayor de 65 años
elif edad < 12 or edad > 65:
    prioridad, nivel, emoji_prioridad = "AMARILLA", "Estándar", "🟡"

# 4. Prioridad VERDE: cualquier otro caso
else:
    prioridad, nivel, emoji_prioridad = "VERDE", "No urgente", "🟢"


# ============================================================
#   PROTOCOLO DE BIOSEGURIDAD (operador not)
# ============================================================

# Box de Aislamiento: síntomas infecciosos AND NOT vacunado
if tiene_sintomas_infecciosos and not esta_vacunado:
    destino, emoji_destino = "BOX DE AISLAMIENTO", "🚨"
else:
    destino, emoji_destino = "SALA DE ESPERA GENERAL", "🪑"


# ============================================================
#   RESUMEN FINAL DEL PACIENTE
# ============================================================

sep = "=" * 45

print(sep)
print("       🏥  INFORME DE TRIAJE")
print(sep)
print(f"  Edad               : {edad} años")
print(f"  Temperatura        : {temperatura} °C")
print(f"  Frecuencia cardíaca: {frecuencia_cardiaca} bpm")
print(f"  Síntomas infecciosos: {'Sí' if tiene_sintomas_infecciosos else 'No'}")
print(f"  Vacunado           : {'Sí' if esta_vacunado else 'No'}")
print(f"  Emergencia crítica : {'Sí' if es_emergencia_critica else 'No'}")
print(sep)
print(f"  {emoji_prioridad}  Prioridad : {prioridad} — {nivel}")
print(f"  {emoji_destino}  Destino   : {destino}")
print(sep)
print("  ✅ Notificar al médico de guardia.")
print(sep)
```
