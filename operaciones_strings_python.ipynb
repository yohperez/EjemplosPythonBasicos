# 🐍 Operaciones con Cadenas (Strings) en Python

---

> **Nota importante:** En Python, las cadenas son **inmutables** — no puedes cambiar una letra dentro de una cadena existente. Todas las operaciones que veremos aquí **crean una nueva cadena** como resultado.

---

Este notebook cubre las principales categorías de operaciones con strings:

| # | Operación | Símbolo / Método |
|---|-----------|------------------|
| 1 | Concatenación | `+` |
| 2 | Repetición | `*` |
| 3 | Indexación y Slicing | `[]` |
| 4 | Métodos de transformación | `.upper()`, `.lower()`, `.strip()`, `.replace()` |

## 1. 🔗 Concatenación (Unión)

Se usa el operador **`+`** para unir dos o más cadenas en una sola.

```
"Hola" + ", " + "Yoh" + "!"  →  "Hola, Yoh!"
```

````python
# --- Concatenación básica ---
saludo = "Hola"
nombre = "Yoh"
mensaje = saludo + ", " + nombre + "!"

print(mensaje)  # Resultado esperado: Hola, Yoh!
````

````python
# --- Pruébalo tú mismo: cambia las variables y observa el resultado ---
ciudad    = "Madrid"
pais      = "España"
resultado = ciudad + " es la capital de " + pais + "."

print(resultado)
````

## 2. 🔁 Repetición

Se usa el operador **`*`** para repetir una cadena un número determinado de veces.

```
"-" * 10  →  "----------"
```

````python
# --- Repetición básica ---
separador = "-" * 10
print(separador)   # ----------

# Útil para crear separadores visuales en la consola
print("=" * 30)
print("   MENÚ PRINCIPAL   ")
print("=" * 30)
````

## 3. ✂️ Indexación y Slicing (Acceso a caracteres)

Las cadenas son **secuencias**: cada carácter tiene una posición (índice) que empieza en **0**.

```
 P  y  t  h  o  n
 0  1  2  3  4  5   ← índices positivos
-6 -5 -4 -3 -2 -1   ← índices negativos
```

- **Indexación:** `cadena[índice]` → un solo carácter
- **Slicing:** `cadena[inicio:fin]` → subcadena (el `fin` **no** se incluye)
- **Slicing con paso:** `cadena[inicio:fin:paso]`

````python
# --- Indexación ---
texto = "Python"

print(texto[0])   # P  ← primer carácter
print(texto[3])   # h
print(texto[-1])  # n  ← último carácter
print(texto[-2])  # o  ← penúltimo
````

````python
# --- Slicing básico ---
texto = "Python"

print(texto[1:4])  # yth  (índices 1, 2 y 3 — el 4 no se incluye)
print(texto[:3])   # Pyt  (desde el inicio hasta el índice 2)
print(texto[3:])   # hon  (desde el índice 3 hasta el final)
````

````python
# --- Slicing con paso ---
texto = "Python"

print(texto[::2])   # Pto  (cada 2 caracteres)
print(texto[::-1])  # nohtyP  ← cadena invertida (truco clásico)
````

````python
# --- Ejercicio guiado ---
frase = "Aprendiendo Python"

# Extrae solo la palabra "Python" usando slicing
# Pista: ¿en qué índice empieza la 'P' de Python?
print(len(frase))           # Longitud total de la cadena
print(frase[12:])           # Extracción

# ¿Cuál es el tercer carácter?
print(frase[2])
````

## 4. 🔧 Métodos útiles (Transformación)

Python incluye métodos integrados que devuelven una **nueva cadena** transformada.

| Método | Descripción |
|--------|-------------|
| `.strip()` | Elimina espacios al inicio y al final |
| `.upper()` | Convierte todo a MAYÚSCULAS |
| `.lower()` | Convierte todo a minúsculas |
| `.replace(a, b)` | Reemplaza todas las ocurrencias de `a` por `b` |
| `.split(sep)` | Divide la cadena en una lista usando `sep` como separador |
| `.join(lista)` | Une los elementos de una lista con la cadena como separador |

````python
# --- strip, upper, lower ---
texto = "  hola mundo  "

print(repr(texto))          # Muestra la cadena con los espacios visibles
print(texto.strip())        # "hola mundo" ← sin espacios al inicio/final
print(texto.upper())        # "  HOLA MUNDO  "
print(texto.lower())        # "  hola mundo  "
print(texto.strip().upper()) # Encadenando métodos: "HOLA MUNDO"
````

````python
# --- replace ---
texto = "  hola mundo  "

nuevo = texto.replace("hola", "adiós")
print(nuevo)   # "  adiós mundo  "

# Reemplazar múltiples veces (encadenando)
frase = "el gato come el ratón"
frase = frase.replace("gato", "perro").replace("ratón", "hueso")
print(frase)
````

````python
# --- split y join ---
csv_line = "manzana,pera,naranja,uva"
frutas = csv_line.split(",")   # ['manzana', 'pera', 'naranja', 'uva']
print(frutas)
print(type(frutas))             # Es una lista

# Volver a unir con un separador diferente
resultado = " | ".join(frutas)
print(resultado)   # manzana | pera | naranja | uva
````

## 5. ✨ Formateo con f-strings

Las **f-strings** son la forma más moderna y legible de insertar variables dentro de un texto.
Solo hay que anteponer una **`f`** antes de las comillas y usar **`{}`** para incrustar expresiones.

```python
f"Texto {variable} más texto"
```

````python
# --- F-string básica ---
nombre = "Yoh"
edad   = 17

print(f"Hola, me llamo {nombre} y tengo {edad} años.")
````

````python
# --- Expresiones dentro de las llaves ---
precio   = 19.5
cantidad = 3

print(f"Total a pagar: {precio * cantidad:.2f} €")  # :.2f → 2 decimales
print(f"El doble de tu edad es {edad * 2}")
print(f"Tu nombre en mayúsculas: {nombre.upper()}")
````

````python
# --- Comparación: formas antiguas vs f-string ---
producto = "portátil"
stock    = 7

# Forma antigua con format()
print("Quedan {} unidades de {}".format(stock, producto))

# Forma moderna con f-string ← recomendada
print(f"Quedan {stock} unidades de {producto}")
````

## 6. 🔍 Verificación de pertenencia (`in`)

El operador **`in`** permite comprobar si una subcadena existe dentro de otra.
Devuelve `True` o `False`.

````python
# --- Uso básico de 'in' ---
texto = "Aprender Python es genial"

print("Python" in texto)   # True
print("Java"   in texto)   # False
print("genial" in texto)   # True

# También se puede usar 'not in'
print("Java" not in texto)  # True
````

````python
# --- Uso con condicionales ---
email = "usuario@ejemplo.com"

if "@" in email:
    print("✅ El email parece válido (contiene '@')")
else:
    print("❌ El email no es válido")
````

## 7. 🔄 Conversión de tipos

A veces los datos vienen como texto (string) pero necesitas operar con ellos como números, o al revés.

| Función | Convierte a... |
|---------|----------------|
| `int(x)` | Entero |
| `float(x)` | Decimal |
| `str(x)` | Cadena de texto |

````python
# --- De string a número ---
numero_str = "100"
numero_int = int(numero_str)

print(numero_int + 50)    # 150  ← ahora sí podemos sumar
print(type(numero_str))   # <class 'str'>
print(type(numero_int))   # <class 'int'>
````

````python
# --- De número a string ---
precio = 29.99
mensaje = "El precio es: " + str(precio) + " €"
print(mensaje)

# Nota: sin str() daría un TypeError
# "El precio es: " + 29.99   ← ERROR en Python
````

````python
# --- Conversión con float ---
temp_str = "36.5"
temp     = float(temp_str)
print(f"La temperatura es {temp} °C")
print(f"En Fahrenheit: {temp * 9/5 + 32:.1f} °F")
````
