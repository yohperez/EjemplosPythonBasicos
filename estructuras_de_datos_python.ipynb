# 🗂️ Estructuras de Datos en Python
## `list` · `tuple` · `set` · `dict`

---

Python ofrece cuatro estructuras de datos fundamentales. Cada una tiene características distintas que la hacen ideal para diferentes situaciones.

| Característica | `list` | `tuple` | `set` | `dict` |
|----------------|:------:|:-------:|:-----:|:------:|
| Sintaxis        | `[1, 2, 3]` | `(1, 2, 3)` | `{1, 2, 3}` | `{"a": 1}` |
| ¿Ordenada?     | ✅ | ✅ | ❌ | ✅ (Python 3.7+) |
| ¿Indexada?     | ✅ | ✅ | ❌ | ✅ (por clave) |
| ¿Duplicados?   | ✅ | ✅ | ❌ | ❌ (claves únicas) |
| ¿Anidable?     | ✅ | ✅ | ❌ | ✅ (valores) |
| ¿Mutable?      | ✅ | ❌ | ✅ | ✅ |
| ¿Dinámica?     | ✅ | ❌ | ✅ | ✅ |
| ¿Iterable?     | ✅ | ✅ | ✅ | ✅ |

## 1. 📋 Listas (`list`)

> **¿Qué son?** Colecciones **ordenadas y mutables**. Son el "todoterreno" de Python.
>
> **Úsalas cuando:** necesites añadir, eliminar o reordenar elementos constantemente.

```python
# ── Creación ──────────────────────────────────────────────
tareas = ["comprar pan", "estudiar Python", "hacer ejercicio"]
numeros = [10, 20, 30, 40, 50]
mixta   = ["Yoh", 44, True, 3.14]   # puede mezclar tipos

print("Lista de tareas:", tareas)
print("Tipo:", type(tareas))
```

```python
# ── Indexación y Slicing ──────────────────────────────────
frutas = ["manzana", "pera", "naranja", "uva", "mango"]

print(frutas[0])     # manzana  ← primer elemento
print(frutas[-1])    # mango    ← último elemento
print(frutas[1:3])   # ['pera', 'naranja']  ← slicing
print(len(frutas))   # 5  ← longitud
```

```python
# ── Mutabilidad: modificar elementos ─────────────────────
frutas[1] = "melón"          # reemplazamos 'pera' por 'melón'
print("Después de modificar:", frutas)
```

```python
# ── Métodos principales ───────────────────────────────────
frutas = ["manzana", "pera", "naranja"]

frutas.append("uva")              # añade al final
print("append('uva'):", frutas)

frutas.insert(1, "kiwi")          # inserta en posición 1
print("insert(1,'kiwi'):", frutas)

eliminado = frutas.pop()          # elimina y devuelve el último
print(f"pop() eliminó '{eliminado}':", frutas)

frutas.sort()                     # ordena alfabéticamente
print("sort():", frutas)

frutas.reverse()                  # invierte el orden
print("reverse():", frutas)
```

```python
# ── Duplicados permitidos ─────────────────────────────────
votos = ["Pedro", "Ana", "Pedro", "Luis", "Ana", "Pedro"]
print("Votos:", votos)
print(f"Pedro recibió {votos.count('Pedro')} votos")

# ── Listas anidadas ───────────────────────────────────────
matriz = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]
print("Elemento [1][2]:", matriz[1][2])  # 6
```

```python
# ── Iteración ─────────────────────────────────────────────
compras = ["leche", "huevos", "pan", "café"]

print("Lista de la compra:")
for i, producto in enumerate(compras, start=1):
    print(f"  {i}. {producto}")
```

## 2. 📌 Tuplas (`tuple`)

> **¿Qué son?** Colecciones **ordenadas e inmutables**. Una vez creadas, no cambian.
>
> **Úsalas cuando:** los datos no deben modificarse: coordenadas GPS, configuraciones fijas, valores constantes.

```python
# ── Creación ──────────────────────────────────────────────
coordenadas  = (40.4168, -3.7038)          # Madrid
colores_rgb  = (255, 128, 0)               # naranja
config_bd    = ("localhost", 5432, "mydb") # host, puerto, nombre

print("Coordenadas Madrid:", coordenadas)
print("Tipo:", type(coordenadas))
```

```python
# ── Indexación (igual que las listas) ────────────────────
punto = (10, 20, 30)

print("x =", punto[0])
print("y =", punto[1])
print("z =", punto[2])
print("Últimos dos:", punto[-2:])
```

```python
# ── Inmutabilidad: no se puede modificar ─────────────────
coordenadas = (40.4168, -3.7038)

try:
    coordenadas[0] = 99   # ← intentamos cambiar el valor
except TypeError as e:
    print(f"❌ Error: {e}")
    print("Las tuplas son inmutables: no se pueden modificar.")
```

```python
# ── Métodos disponibles ───────────────────────────────────
notas = (8, 9, 7, 10, 9, 8, 9)

print("Apariciones del 9:", notas.count(9))   # count()
print("Posición del 10:",  notas.index(10))   # index()
print("Máxima nota:",      max(notas))
print("Media:",            sum(notas) / len(notas))
```

```python
# ── Desempaquetado (unpacking) ────────────────────────────
# Una ventaja muy práctica de las tuplas
latitud, longitud = (40.4168, -3.7038)
print(f"Latitud: {latitud}  |  Longitud: {longitud}")

# También en bucles
ciudades = [("Madrid", 3.3), ("Barcelona", 1.6), ("Valencia", 0.8)]
for ciudad, millones in ciudades:
    print(f"{ciudad}: {millones}M habitantes")
```

## 3. 🔵 Sets (`set`)

> **¿Qué son?** Colecciones **no ordenadas** de elementos **únicos**.
>
> **Úsalos cuando:** necesites eliminar duplicados o realizar operaciones matemáticas de conjuntos.

```python
# ── Creación ──────────────────────────────────────────────
colores = {"rojo", "verde", "azul", "rojo", "verde"}  # duplicados ignorados
print("Set de colores:", colores)
print("Tipo:", type(colores))
print("⚠️  Los duplicados se eliminaron automáticamente")
```

```python
# ── Sin índices: no se puede acceder por posición ────────
colores = {"rojo", "verde", "azul"}

try:
    print(colores[0])   # ← esto falla
except TypeError as e:
    print(f"❌ Error: {e}")
    print("Los sets no tienen índices.")

# Pero sí podemos comprobar pertenencia:
print("¿'azul' en colores?", "azul" in colores)
```

```python
# ── Métodos: add y remove ─────────────────────────────────
lenguajes = {"Python", "JavaScript", "Java"}

lenguajes.add("Rust")           # añade un elemento
print("Después de add('Rust'):", lenguajes)

lenguajes.discard("Java")       # elimina si existe (sin error si no existe)
print("Después de discard('Java'):", lenguajes)
```

```python
# ── Operaciones de conjuntos ──────────────────────────────
grupo_a = {"Ana", "Luis", "Marta", "Pedro"}
grupo_b = {"Luis", "Carmen", "Pedro", "Rosa"}

print("Unión (todos):",          grupo_a | grupo_b)          # union()
print("Intersección (comunes):", grupo_a & grupo_b)          # intersection()
print("Diferencia (solo en A):", grupo_a - grupo_b)          # difference()
print("Diferencia simétrica:",   grupo_a ^ grupo_b)          # symmetric_difference()
```

```python
# ── Caso de uso clásico: eliminar duplicados de una lista ─
visitas = ["Madrid", "París", "Madrid", "Roma", "París", "Tokio", "Roma"]
print("Lista original:", visitas)
print("Ciudades únicas:", list(set(visitas)))
print(f"Visitaste {len(set(visitas))} ciudades distintas")
```

## 4. 📖 Diccionarios (`dict`)

> **¿Qué son?** Colecciones de pares **clave:valor**. Muy rápidos para buscar datos.
>
> **Úsalos cuando:** necesites asociar datos con un nombre o identificador (como una base de datos).

```python
# ── Creación ──────────────────────────────────────────────
usuario = {
    "nombre": "Yoh",
    "edad":   44,
    "ciudad": "Madrid",
    "activo": True
}

print(usuario)
print("Tipo:", type(usuario))
```

```python
# ── Acceso por clave ──────────────────────────────────────
print(usuario["nombre"])          # acceso directo
print(usuario.get("edad"))        # acceso seguro con .get()
print(usuario.get("email", "N/A"))# valor por defecto si no existe
```

```python
# ── Mutabilidad: añadir, modificar y eliminar ─────────────
usuario["email"] = "yoh@ejemplo.com"  # añadir nueva clave
usuario["edad"]  = 45                 # modificar valor existente
del usuario["activo"]                 # eliminar clave

print(usuario)
```

```python
# ── Métodos principales ───────────────────────────────────
producto = {"nombre": "Portátil", "precio": 899.99, "stock": 15}

print("Claves:",  list(producto.keys()))
print("Valores:", list(producto.values()))
print("Items:",   list(producto.items()))
```

```python
# ── Iteración ─────────────────────────────────────────────
persona = {"nombre": "Ana", "profesión": "Ingeniera", "ciudad": "Sevilla"}

print("── Solo claves ──")
for clave in persona:
    print(f"  {clave}")

print("── Clave y valor ──")
for clave, valor in persona.items():
    print(f"  {clave}: {valor}")
```

```python
# ── Claves únicas (sin duplicados) ────────────────────────
config = {"host": "localhost", "puerto": 8080, "host": "produccion.com"}
# La segunda definición de 'host' sobreescribe la primera
print(config)   # 'host' → 'produccion.com'
```

```python
# ── Diccionarios anidados ─────────────────────────────────
empleados = {
    "E001": {"nombre": "Carlos", "dept": "IT",      "salario": 3200},
    "E002": {"nombre": "Laura",  "dept": "Marketing","salario": 2900},
    "E003": {"nombre": "Sergio", "dept": "IT",      "salario": 3500},
}

# Acceso a datos anidados
print("Nombre E002:", empleados["E002"]["nombre"])

# Filtrar empleados de IT
print("\nEmpleados de IT:")
for id_emp, datos in empleados.items():
    if datos["dept"] == "IT":
        print(f"  {id_emp} → {datos['nombre']} (€{datos['salario']})")
```

## 5. 🔄 Iteración en todas las estructuras

> Todas las estructuras son **iterables**: puedes usar `for` en cualquiera de ellas.

```python
mi_lista = ["a", "b", "c"]
mi_tupla = (10, 20, 30)
mi_set   = {"rojo", "verde", "azul"}
mi_dict  = {"x": 1, "y": 2, "z": 3}

print("Lista:  ", end=""); [print(x, end=" ") for x in mi_lista]; print()
print("Tupla:  ", end=""); [print(x, end=" ") for x in mi_tupla]; print()
print("Set:    ", end=""); [print(x, end=" ") for x in mi_set];   print()
print("Dict (claves):", end=" "); [print(k, end=" ") for k in mi_dict]; print()
print("Dict (valores):",end=" "); [print(v, end=" ") for v in mi_dict.values()]; print()
```

## 6. ⚖️ ¿Cuándo usar cada una?

| Necesidad | Estructura ideal | ¿Por qué? |
|-----------|:----------------:|-----------|
| Colección que cambia constantemente | `list` | Mutable y dinámica |
| Datos que NO deben cambiar | `tuple` | Inmutable, más segura y rápida |
| Eliminar duplicados / unicidad | `set` | Garantiza elementos únicos |
| Buscar por nombre o ID | `dict` | Acceso O(1) por clave |
| Operaciones matemáticas de conjuntos | `set` | Unión, intersección, diferencia |
| Representar un objeto con atributos | `dict` | Par clave:valor |

## 🧪 Ejercicio Final: Gestión de una clase

Usa las **cuatro estructuras** para gestionar datos de una clase de Python.

```python
# ── 1. LISTA: alumnos inscritos (puede cambiar) ───────────
alumnos = ["Ana", "Luis", "Marta", "Pedro", "Carmen", "Luis"]  # Luis duplicado
alumnos.append("Rosa")
print(f"Alumnos inscritos ({len(alumnos)}): {alumnos}")

# ── 2. SET: alumnos únicos (sin duplicados) ───────────────
alumnos_unicos = set(alumnos)
print(f"Alumnos únicos ({len(alumnos_unicos)}): {alumnos_unicos}")

# ── 3. TUPLA: datos del curso (no cambian) ────────────────
info_curso = ("Python para principiantes", "2025", "Aula 3", 20)
nombre_curso, año, aula, max_alumnos = info_curso   # unpacking
print(f"\nCurso: {nombre_curso} | Año: {año} | {aula} | Máx: {max_alumnos}")

# ── 4. DICCIONARIO: notas por alumno ─────────────────────
notas = {
    "Ana":    9.5,
    "Luis":   7.0,
    "Marta":  8.5,
    "Pedro":  6.0,
    "Carmen": 9.0,
    "Rosa":   8.0,
}

print("\n" + "═" * 35)
print(f"{'ALUMNO':<12} {'NOTA':>6} {'RESULTADO':>10}")
print("═" * 35)
for alumno, nota in notas.items():
    resultado = "✅ Aprobado" if nota >= 5 else "❌ Suspenso"
    print(f"{alumno:<12} {nota:>6.1f} {resultado:>10}")

media = sum(notas.values()) / len(notas)
mejor = max(notas, key=notas.get)
print("═" * 35)
print(f"Media de clase: {media:.2f}")
print(f"Mejor nota: {mejor} ({notas[mejor]})")
```

## 📋 Resumen de Métodos Clave

### `list`
| Método | Descripción |
|--------|-------------|
| `.append(x)` | Añade `x` al final |
| `.insert(i, x)` | Inserta `x` en la posición `i` |
| `.pop()` | Elimina y devuelve el último elemento |
| `.sort()` | Ordena la lista en su lugar |
| `.reverse()` | Invierte la lista en su lugar |
| `.count(x)` | Cuenta las apariciones de `x` |

### `tuple`
| Método | Descripción |
|--------|-------------|
| `.count(x)` | Cuenta las apariciones de `x` |
| `.index(x)` | Devuelve la posición de `x` |

### `set`
| Método | Descripción |
|--------|-------------|
| `.add(x)` | Añade `x` al set |
| `.discard(x)` | Elimina `x` (sin error si no existe) |
| `.union(s)` / `\|` | Todos los elementos de ambos |
| `.intersection(s)` / `&` | Solo los elementos comunes |
| `.difference(s)` / `-` | Elementos en A pero no en B |

### `dict`
| Método | Descripción |
|--------|-------------|
| `.get(k, def)` | Devuelve valor de `k` o `def` si no existe |
| `.keys()` | Vista de todas las claves |
| `.values()` | Vista de todos los valores |
| `.items()` | Vista de pares (clave, valor) |
| `.update(d)` | Fusiona con otro diccionario `d` |
