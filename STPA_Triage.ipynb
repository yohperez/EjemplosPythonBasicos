# 🏥 Sistema de Triaje y Protocolo de Aislamiento (STPA)

**Contexto:** Software para personal de enfermería que clasifica pacientes según su gravedad física y determina si deben ser aislados por riesgo biológico.

---

## 📋 Variables de entrada

| Variable | Tipo | Descripción |
|---|---|---|
| `edad` | `int` | Edad del paciente en años |
| `frecuencia_cardiaca` | `int` | Pulsaciones por minuto |
| `temperatura` | `float` | Temperatura corporal en °C |
| `tiene_sintomas_infecciosos` | `bool` | Presencia de síntomas infecciosos |
| `esta_vacunado` | `bool` | Estado de vacunación del paciente |
| `es_emergencia_critica` | `bool` | Indicador de emergencia crítica |

---

## 1️⃣ Ingreso de datos del paciente

Introduce los valores del paciente en las variables de la celda siguiente.

```python
# ============================================================
#   DATOS DEL PACIENTE — Modifica estos valores
# ============================================================

edad                     = 70        # Entero: años del paciente
frecuencia_cardiaca      = 105       # Entero: pulsaciones por minuto
temperatura              = 39.5      # Flotante: temperatura en °C
tiene_sintomas_infecciosos = True    # Boolean: True / False
esta_vacunado            = False     # Boolean: True / False
es_emergencia_critica    = False     # Boolean: True / False

print("✅ Datos del paciente registrados correctamente.")
```

---

## 2️⃣ Clasificación de Prioridad

El sistema evalúa las condiciones en **orden estricto** de prioridad:

1. 🔴 **Roja (Inmediata):** emergencia crítica *o* temperatura > 41 °C  
2. 🟠 **Naranja (Urgente):** temperatura ≥ 39 °C *y* frecuencia cardíaca > 100  
3. 🟡 **Amarilla (Estándar):** menor de 12 años *o* mayor de 65 años  
4. 🟢 **Verde (No urgente):** cualquier otro caso

```python
# ============================================================
#   MÓDULO 1: Clasificación de Prioridad (if / elif / else)
# ============================================================

# Condición 1 — Prioridad ROJA: emergencia crítica O temperatura muy alta
if es_emergencia_critica or temperatura > 41:
    prioridad       = "ROJA"
    nivel           = "Inmediata"
    emoji_prioridad = "🔴"

# Condición 2 — Prioridad NARANJA: fiebre alta Y taquicardia
elif temperatura >= 39 and frecuencia_cardiaca > 100:
    prioridad       = "NARANJA"
    nivel           = "Urgente"
    emoji_prioridad = "🟠"

# Condición 3 — Prioridad AMARILLA: paciente pediátrico O adulto mayor
elif edad < 12 or edad > 65:
    prioridad       = "AMARILLA"
    nivel           = "Estándar"
    emoji_prioridad = "🟡"

# Condición 4 — Prioridad VERDE: cualquier otro caso
else:
    prioridad       = "VERDE"
    nivel           = "No urgente"
    emoji_prioridad = "🟢"

print(f"Prioridad asignada: {emoji_prioridad} {prioridad} ({nivel})")
```

---

## 3️⃣ Protocolo de Bioseguridad

Determina el **destino del paciente** de forma independiente a la prioridad:

- 🚨 **Box de Aislamiento:** síntomas infecciosos **y** paciente **no** vacunado  
- 🪑 **Sala de Espera General:** sin síntomas infecciosos, o vacunado aunque tenga síntomas

```python
# ============================================================
#   MÓDULO 2: Protocolo de Bioseguridad (operador not)
# ============================================================

# Box de Aislamiento: síntomas infecciosos AND NOT vacunado
if tiene_sintomas_infecciosos and not esta_vacunado:
    destino       = "BOX DE AISLAMIENTO"
    emoji_destino = "🚨"
    razon_destino = "Síntomas infecciosos activos sin vacunación — riesgo biológico elevado"

# Sala General: sin síntomas, o síntomas pero vacunado
else:
    destino       = "SALA DE ESPERA GENERAL"
    emoji_destino = "🪑"
    razon_destino = "Sin riesgo biológico significativo"

print(f"Destino asignado: {emoji_destino} {destino}")
```

---

## 4️⃣ Resumen Final del Paciente

```python
# ============================================================
#   RESUMEN FINAL — Informe del paciente
# ============================================================

separador = "=" * 55

print(separador)
print("       🏥  SISTEMA DE TRIAJE Y PROTOCOLO DE AISLAMIENTO")
print(separador)

print("\n📊  DATOS CLÍNICOS DEL PACIENTE")
print(f"   • Edad               : {edad} años")
print(f"   • Temperatura        : {temperatura} °C")
print(f"   • Frecuencia cardíaca: {frecuencia_cardiaca} bpm")
print(f"   • Síntomas infecciosos: {'Sí' if tiene_sintomas_infecciosos else 'No'}")
print(f"   • Vacunado           : {'Sí' if esta_vacunado else 'No'}")
print(f"   • Emergencia crítica : {'Sí' if es_emergencia_critica else 'No'}")

print("\n" + separador)

print("\n🎯  RESULTADO DEL TRIAJE")
print(f"   {emoji_prioridad}  Prioridad : {prioridad} — {nivel}")
print(f"   {emoji_destino}  Destino   : {destino}")
print(f"   📝  Motivo    : {razon_destino}")

print("\n" + separador)
print("   ✅  Clasificación completada. Notificar al médico de guardia.")
print(separador)
```

---

## 5️⃣ Prueba con múltiples casos

Celda de **prueba automatizada** con 5 pacientes de ejemplo para verificar que la lógica funciona correctamente en todos los escenarios.

```python
# ============================================================
#   FUNCIÓN STPA — Encapsula toda la lógica del sistema
# ============================================================

def clasificar_paciente(edad, frecuencia_cardiaca, temperatura,
                         tiene_sintomas_infecciosos, esta_vacunado,
                         es_emergencia_critica):
    """
    Clasifica a un paciente según el protocolo STPA.

    Parámetros:
        edad (int)                        : Edad del paciente.
        frecuencia_cardiaca (int)         : Pulsaciones por minuto.
        temperatura (float)               : Temperatura corporal en °C.
        tiene_sintomas_infecciosos (bool) : Presencia de síntomas infecciosos.
        esta_vacunado (bool)              : Estado de vacunación.
        es_emergencia_critica (bool)      : Indicador de emergencia crítica.

    Retorna:
        tuple: (prioridad, nivel, destino, razon_destino)
    """

    # --- Clasificación de prioridad ---
    if es_emergencia_critica or temperatura > 41:
        prioridad, nivel = "ROJA",     "Inmediata"
    elif temperatura >= 39 and frecuencia_cardiaca > 100:
        prioridad, nivel = "NARANJA",  "Urgente"
    elif edad < 12 or edad > 65:
        prioridad, nivel = "AMARILLA", "Estándar"
    else:
        prioridad, nivel = "VERDE",    "No urgente"

    # --- Protocolo de bioseguridad ---
    if tiene_sintomas_infecciosos and not esta_vacunado:
        destino       = "BOX DE AISLAMIENTO"
        razon_destino = "Riesgo biológico — sin vacunar"
    else:
        destino       = "SALA DE ESPERA GENERAL"
        razon_destino = "Sin riesgo biológico significativo"

    return prioridad, nivel, destino, razon_destino


# ============================================================
#   CASOS DE PRUEBA
# ============================================================

casos = [
    # (edad, fc, temp, sintomas, vacunado, critica,  descripcion)
    (35,  90, 38.0, False, True,  False, "Adulto sano, sin síntomas"),
    (45, 110, 39.8, True,  False, False, "Fiebre alta + taquicardia + sin vacuna"),
    (8,   80, 37.5, False, True,  False, "Niño sin síntomas"),
    (72,  95, 37.0, True,  True,  False, "Adulto mayor vacunado con síntomas"),
    (30,  75, 42.1, False, True,  False, "Temperatura crítica > 41 °C"),
]

emojis = {"ROJA": "🔴", "NARANJA": "🟠", "AMARILLA": "🟡", "VERDE": "🟢"}

print("=" * 65)
print("  PRUEBA AUTOMATIZADA — 5 CASOS DE PACIENTE")
print("=" * 65)

for i, (edad, fc, temp, sint, vac, crit, desc) in enumerate(casos, 1):
    prioridad, nivel, destino, razon = clasificar_paciente(
        edad, fc, temp, sint, vac, crit
    )
    print(f"\nCaso {i}: {desc}")
    print(f"  {emojis[prioridad]} Prioridad : {prioridad} ({nivel})")
    print(f"  📍 Destino   : {destino}")
    print(f"  📝 Motivo    : {razon}")

print("\n" + "=" * 65)
print("  ✅  Todos los casos procesados.")
print("=" * 65)
```

| Operador `and` | Prioridad naranja: `temperatura >= 39 and frecuencia_cardiaca > 100` |
| Operador `not` | Box de aislamiento: `tiene_sintomas_infecciosos and not esta_vacunado` |
| Variables `bool` | `tiene_sintomas_infecciosos`, `esta_vacunado`, `es_emergencia_critica` |
| `f-strings` | Impresión del resumen final |
| Función con `def` | Encapsulación del sistema en la celda de prueba |
